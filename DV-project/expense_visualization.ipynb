```python
# --- INITIALIZATION ---

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter
import ipywidgets as widgets
from ipywidgets import interact_manual, HBox, HTML, VBox, Button
from IPython.display import display, clear_output
import calendar
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=pd.errors.SettingWithCopyWarning)

plt.style.use('ggplot')
sns.set_palette("Set2")
```

```python
# --- Formatter ---
def inr(x, _): return f"â‚¹{x:,.0f}"
currency_fmt = FuncFormatter(inr)
```

```python
# Normalize all paid_by entries
for data in (house_data, studio_data):
    for row in data:
        row['paid_by'] = row['paid_by'].strip().title().replace('  ', ' ')
```

```python
# Convert to DataFrame
house_df = pd.DataFrame(house_data)
studio_df = pd.DataFrame(studio_data)

for df in (house_df, studio_df):
    df['date'] = pd.to_datetime(df['date'])
    df['month'] = df['date'].dt.month_name()
    df['day'] = df['date'].dt.day_name()

summary_df = pd.concat([
    house_df.assign(Project='House'),
    studio_df.assign(Project='Studio')
])
```

```python
# --- Chart Function ---
def plot_chart(df, group, chart_type, title):
    agg = df.groupby(group)['amount'].sum().sort_values(ascending=False)

    if chart_type == "Bar Chart":
        plt.figure()
        ax = agg.plot(kind='bar', figsize=(max(9, len(agg) * 0.5), 4))
        ax.yaxis.set_major_formatter(currency_fmt)
        plt.title(title)
        plt.tight_layout()
        plt.show()

    elif chart_type == "Pie Chart":
        plt.figure()
        plt.pie(agg, labels=agg.index, autopct='%1.1f%%', startangle=140)
        plt.axis('equal')
        plt.title(title)
        plt.tight_layout()
        plt.show()

    elif chart_type == "Trend Line":
        trend_df = df.groupby('date')['amount'].sum()
        trend_df.plot(marker='o')
        plt.gca().yaxis.set_major_formatter(currency_fmt)
        plt.title("Spending Over Time")
        plt.xlabel("Date")
        plt.ylabel("Amount")
        plt.grid(True)
        plt.tight_layout()
        plt.show()

    elif chart_type == "Top Contributors":
        top_df = df.groupby('paid_by')['amount'].sum().sort_values(ascending=False).head(10)
        top_df.plot(kind='bar')
        plt.title("Top Contributors")
        plt.gca().yaxis.set_major_formatter(currency_fmt)
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

    elif chart_type == "Heatmap":
        heat_df = df.groupby(['day', 'month'])['amount'].sum().unstack(fill_value=0)
        plt.figure(figsize=(12, 6))
        sns.heatmap(heat_df, annot=True, fmt='.0f', cmap='YlGnBu')
        plt.title("Weekly Heatmap")
        plt.tight_layout()
        plt.show()

    elif chart_type == "Summary Table":
        summary = df.groupby(group)['amount'].sum().reset_index().sort_values(by='amount', ascending=False)
        styled = summary.style.format({'amount': 'â‚¹{:,.0f}'}).set_caption("Summary Table")
        display(styled)
```

```python
# --- Show Raw Data ---
def show_data_table():
    if summary_df.empty:
        display(HTML("<p style='color:red;'>No data available.</p>"))
    else:
        display(HTML("<h4>Raw Data Table</h4>"))
        display(summary_df.sort_values(by='date').reset_index(drop=True).style.format({'amount': 'â‚¹{:,.0f}'}))
```

```python
# --- Dashboard ---
def run_dashboard():
    chart_selector = widgets.Dropdown(
        options=['Bar Chart', 'Pie Chart', 'Trend Line', 'Top Contributors', 'Heatmap', 'Summary Table'],
        value='Bar Chart', description='Chart Type:'
    )
    group_selector = widgets.Dropdown(
        options=['given_to', 'paid_by', 'payment_mode', 'day', 'month'],
        value='month', description='Group By:'
    )
    month_selector = widgets.Dropdown(
        options=['All'] + sorted(summary_df['month'].unique()),
        value='All', description='Month:'
    )
    mode_selector = widgets.Dropdown(
        options=['All'] + sorted(summary_df['payment_mode'].dropna().unique()),
        value='All', description='Mode:'
    )
    source_selector = widgets.Dropdown(
        options=['All'] + sorted(summary_df['Project'].unique()),
        value='All', description='Source:'
    )

    output = widgets.Output()

    def filter_data():
        df = summary_df.copy()
        if month_selector.value != 'All':
            df = df[df['month'] == month_selector.value]
        if mode_selector.value != 'All':
            df = df[df['payment_mode'] == mode_selector.value]
        if source_selector.value != 'All':
            df = df[df['Project'] == source_selector.value]
        return df

    def update_chart(change):
        with output:
            clear_output(wait=True)
            df = filter_data()
            plot_chart(df, group_selector.value, chart_selector.value, 'Filtered View')

    for widget in [chart_selector, group_selector, month_selector, mode_selector, source_selector]:
        widget.observe(update_chart, names='value')

    display(VBox([
        HBox([chart_selector, group_selector]),
        HBox([month_selector, mode_selector, source_selector]),
        output
    ]))

    update_chart(None)
```

```python
# --- Dashboard ---
if not summary_df.empty:
    display(HTML("<h3>Total Spending Summary</h3>"))
    total_summary = summary_df.groupby('Project')['amount'].sum().reset_index()
    total_value = summary_df['amount'].sum()
    display(total_summary.style.format({'amount': 'â‚¹{:,.0f}'}).set_caption("Total by Project"))
    display(HTML(f"<h4>Grand Total: â‚¹{total_value:,.0f}</h4>"))

    display(HTML("<hr><h3>Interactive Dashboard</h3>"))
    run_dashboard()
else:
    print("No data loaded. Please verify house_data and studio_data.")
```
